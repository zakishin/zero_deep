```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def And(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.7
    tmp=np.sum(x*w)+b
    if tmp<=0:
        return 0
    else:
        return 1
def nand(x1,x2):
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5])
    b=0.7
    tmp=np.sum(x*w)+b
    if tmp<=0:
        return 0
    else:
        return 1
def OR(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.2
    tmp=np.sum(x*w)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
x=np.arange(-0.5,1.5,0.1)
AND=-x+1.4
NAND=-x+1.4
OR=-x+0.4
plt.plot(x,AND,label='and')
plt.plot(x,NAND,label='nand')
plt.plot(x,OR,label='or')
plt.scatter(1,1)
plt.scatter(0,1)
plt.scatter(1,0)
plt.scatter(0,0)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.legend()
plt.show()
```
